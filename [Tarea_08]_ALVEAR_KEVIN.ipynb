## Resolución del Ejercicio 1

Dados los datos:

$$x_i = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]$$
$$y_i = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]$$

**a. Construir el polinomio de grado 1 por mínimos cuadrados:**

Sea el polinomio de grado 1: $y = a + bx$

Aplicando el método de mínimos cuadrados, tenemos:

$$b = \frac{n\sum_{i=1}^n x_iy_i - \sum_{i=1}^n x_i\sum_{i=1}^n y_i}{n\sum_{i=1}^n x_i^2 - \left(\sum_{i=1}^n x_i\right)^2}$$
$$a = \frac{\sum_{i=1}^n y_i - b\sum_{i=1}^n x_i}{n}$$

Reemplazando los valores dados, obtenemos:

$$b = \frac{10(14,495.23) - (49.6)(1,732.39)}{10(49.6^2) - (49.6)^2} = 32.9$$
$$a = \frac{1,732.39 - 32.9(49.6)}{10} = 12.23$$

Por lo tanto, el polinomio de grado 1 por mínimos cuadrados es:
$y = 12.23 + 32.9x$

**b. Construir el polinomio de grado 2 por mínimos cuadrados:**

Sea el polinomio de grado 2: $y = a + bx + cx^2$

Aplicando el método de mínimos cuadrados, tenemos:

$$c = \frac{n\sum_{i=1}^n x_i^2y_i - \sum_{i=1}^n x_i^2\sum_{i=1}^n y_i - b\left(n\sum_{i=1}^n x_i^3 - \sum_{i=1}^n x_i\sum_{i=1}^n x_i^2\right)}{n\sum_{i=1}^n x_i^4 - \left(\sum_{i=1}^n x_i^2\right)^2}$$
$$b = \frac{\sum_{i=1}^n x_i^2y_i - c\sum_{i=1}^n x_i^3 - a\sum_{i=1}^n x_i^2}{\sum_{i=1}^n x_i^2y_i - c\sum_{i=1}^n x_i^3 - a\sum_{i=1}^n x_i^2}$$
$$a = \frac{\sum_{i=1}^n y_i - b\sum_{i=1}^n x_i - c\sum_{i=1}^n x_i^2}{n}$$

Reemplazando los valores dados, obtenemos:

$$c = \frac{10(18,163.41) - (49.6)(1,732.39) - 32.9(10(2,460.96) - 49.6(49.6))}{10(3,074.56) - (2,460.96)^2} = 1.28$$
$$b = \frac{18,163.41 - 1.28(2,460.96) - 12.23(49.6)}{2,460.96 - 1.28(49.6) - 12.23(1)} = 26.28$$
$$a = \frac{1,732.39 - 26.28(49.6) - 1.28(49.6^2)}{10} = -1.26$$

Por lo tanto, el polinomio de grado 2 por mínimos cuadrados es:
$y = -1.26 + 26.28x + 1.28x^2$

**c. Construir el polinomio de grado 3 por mínimos cuadrados:**

Sea el polinomio de grado 3: $y = a + bx + cx^2 + dx^3$

Aplicando el método de mínimos cuadrados, tenemos:

$$d = \frac{n\sum_{i=1}^n x_i^3y_i - \sum_{i=1}^n x_i^3\sum_{i=1}^n y_i - b\left(n\sum_{i=1}^n x_i^4 - \sum_{i=1}^n x_i^2\sum_{i=1}^n x_i^2\right) - c\left(n\sum_{i=1}^n x_i^5 - \sum_{i=1}^n x_i^2\sum_{i=1}^n x_i^3\right)}{n\sum_{i=1}^n x_i^6 - \left(\sum_{i=1}^n x_i^3\right)^2}$$
$$c = \frac{\sum_{i=1}^n x_i^2y_i - d\sum_{i=1}^n x_i^4 - b\sum_{i=1}^n x_i^3 - a\sum_{i=1}^n x_i^2}{\sum_{i=1}^n x_i^2y_i - d\sum_{i=1}^n x_i^4 - b\sum_{i=1}^n x_i^3 - a\sum_{i=1}^n x_i^2}$$
$$b = \frac{\sum_{i=1}^n x_iy_i - d\sum_{i=1}^n x_i^3 - c\sum_{i=1}^n x_i^2 - a\sum_{i=1}^n x_i}{\sum_{i=1}^n x_iy_i - d\sum_{i=1}^n x_i^3 - c\sum_{i=1}^n x_i^2 - a\sum_{i=1}^n x_i}$$
$$a = \frac{\sum_{i=1}^n y_i - b\sum_{i=1}^n x_i - c\sum_{i=1}^n x_i^2 - d\sum_{i=1}^n x_i^3}{n}$$

Reemplazando los valores dados, obtenemos:

$$d = \frac{10(24,501.87) - (49.6)(1,732.39) - 26.28(10(3,074.56) - 49.6(2,460.96)) - 1.28(10(3,840.96) - 2,460.96(49.6))}{10(4,808.96) - (2,460.96)^2} = 0.23$$
$$c = \frac{18,163.41 - 0.23(3,840.96) - 26.28(2,460.96) - (-1.26)(49.6)}{2,460.96 - 0.23(3,840.96) - 26.28(49.6) - (-1.26)(1)} = 1.12$$
$$b = \frac{14,495.23 - 0.23(2,460.96) - 1.12(49.6) - (-1.26)(1)}{49.6 - 0.23(2,460.96) - 1.12(1) - (-1.26)(49.6)} = 24.85$$
$$a = \frac{1,732.39 - 24.85(49.6) - 1.12(49.6^2) - 0.23(49.6^3)}{10} = -3.89$$

Por lo tanto, el polinomio de grado 3 por mínimos cuadrados es:
$y = -3.89 + 24.85x + 1.12x^2 + 0.23x^3$

**d. Construir el polinomio de la forma $y = b e^{ax}$ por mínimos cuadrados:**

Aplicando el método de mínimos cuadrados, tenemos:

$$a = \frac{n\sum_{i=1}^n x_iy_i - \sum_{i=1}^n x_i\sum_{i=1}^n y_i}{n\sum_{i=1}^n x_i^2 - \left(\sum_{i=1}^n x_i\right)^2}$$
$$b = \frac{\sum_{i=1}^n y_i e^{-ax_i}}{n}$$

Reemplazando los valores dados, obtenemos:

$$a = \frac{10(14,495.23) - (49.6)(1,732.39)}{10(49.6^2) - (49.6)^2} = 0.0636$$
$$b = \frac{1,732.39 e^{-0.0636(49.6)}}{10} = 102.27$$

Por lo tanto, el polinomio de la forma $y = b e^{ax}$ por mínimos cuadrados es:
$y = 102.27 e^{0.0636x}$

**e. Construir el polinomio de la forma $y = bx^a$ por mínimos cuadrados:**

Aplicando el método de mínimos cuadrados, tenemos:

$$a = \frac{n\sum_{i=1}^n \log x_i \log y_i - \sum_{i=1}^n \log x_i \sum_{i=1}^n \log y_i}{n\sum_{i=1}^n (\log x_i)^2 - \left(\sum_{i=1}^n \log x_i\right)^2}$$
$$b = \frac{\sum_{i=1}^n y_i x_i^{-a}}{n}$$

Reemplazando los valores dados, obtenemos:

$$a = \frac{10(1,732.39) - (49.6)(1,732.39)}{10(49.6^2) - (49.6)^2} = 2.1$$
$$b = \frac{1,732.39 (49.6)^{-2.1}}{10} = 0.0002$$

Por lo tanto, el polinomio de la forma $y = bx^a$ por mínimos cuadrados es:
$y = 0.0002 x^{2.1}$

```python
# Ejercicio 1
print("Ejercicio 1:")
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])

def calcular_error(y_real, y_pred):
    return np.sum((y_real - y_pred)**2)

# a. Polinomio de grado 1 (regresión lineal)
coef1 = np.polyfit(x, y, 1)
p1 = np.poly1d(coef1)
error1 = calcular_error(y, p1(x))

# b. Polinomio de grado 2
coef2 = np.polyfit(x, y, 2)
p2 = np.poly1d(coef2)
error2 = calcular_error(y, p2(x))

# c. Polinomio de grado 3
coef3 = np.polyfit(x, y, 3)
p3 = np.poly1d(coef3)
error3 = calcular_error(y, p3(x))

# d. Función de la forma b*e^(ax)
def func_exp(x, a, b):
    return b * np.exp(a * x)

popt_exp, _ = curve_fit(func_exp, x, y)
a_exp, b_exp = popt_exp
error_exp = calcular_error(y, func_exp(x, *popt_exp))

# e. Función de la forma b*x^a
def func_pow(x, a, b):
    return b * x**a

popt_pow, _ = curve_fit(func_pow, x, y)
a_pow, b_pow = popt_pow
error_pow = calcular_error(y, func_pow(x, *popt_pow))

# Visualización
plt.figure(figsize=(12, 8))
plt.scatter(x, y, color='red', label='Datos originales')
x_range = np.linspace(x.min(), x.max(), 100)
plt.plot(x_range, p1(x_range), label=f'Grado 1 (Error: {error1:.2f})')
plt.plot(x_range, p2(x_range), label=f'Grado 2 (Error: {error2:.2f})')
plt.plot(x_range, p3(x_range), label=f'Grado 3 (Error: {error3:.2f})')
plt.plot(x_range, func_exp(x_range, *popt_exp), label=f'Exponencial (Error: {error_exp:.2f})')
plt.plot(x_range, func_pow(x_range, *popt_pow), label=f'Potencial (Error: {error_pow:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por mínimos cuadrados - Ejercicio 1')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir resultados
print(f"Polinomio de grado 1: {p1}")
print(f"Error del polinomio de grado 1: {error1}")

print(f"\nPolinomio de grado 2: {p2}")
print(f"Error del polinomio de grado 2: {error2}")

print(f"\nPolinomio de grado 3: {p3}")
print(f"Error del polinomio de grado 3: {error3}")

print(f"\nFunción exponencial: {b_exp:.4f} * e^({a_exp:.4f}x)")
print(f"Error de la función exponencial: {error_exp}")

print(f"\nFunción potencial: {b_pow:.4f} * x^{a_pow:.4f}")
print(f"Error de la función potencial: {error_pow}")
```

Reemplazando los valores dados, obtenemos:

$$d = \frac{8(2.5015) - (8.6)(2.5015) - 1.3(8(0.256) - 8.6(1.6)) - 0.6(8(0.0512) - 1.6(0.256))}{8(0.0512) - (0.256)^2} = 0.4$$

```python
# Ejercicio 2
print("Ejercicio 2:")
x2 = np.array([0.2, 0.3, 0.6, 0.9, 1.1, 1.3, 1.4, 1.6])
y2 = np.array([0.050446, 0.098426, 0.33277, 0.72660, 1.0972, 1.5697, 1.8487, 2.5015])

def calcular_error(y_real, y_pred):
    return np.sum((y_real - y_pred)**2)

# Polinomios de grado 1, 2 y 3
coefs = [np.polyfit(x2, y2, i) for i in range(1, 4)]
polys = [np.poly1d(coef) for coef in coefs]
errors = [calcular_error(y2, poly(x2)) for poly in polys]

# Función exponencial
def func_exp(x, a, b):
    return b * np.exp(a * x)

popt_exp, _ = curve_fit(func_exp, x2, y2)
error_exp = calcular_error(y2, func_exp(x2, *popt_exp))

# Función potencial
def func_pow(x, a, b):
    return b * x**a

popt_pow, _ = curve_fit(func_pow, x2, y2)
error_pow = calcular_error(y2, func_pow(x2, *popt_pow))

# Visualización
plt.figure(figsize=(12, 8))
plt.scatter(x2, y2, color='red', label='Datos originales')
x_range = np.linspace(x2.min(), x2.max(), 100)
for i, poly in enumerate(polys, 1):
    plt.plot(x_range, poly(x_range), label=f'Grado {i} (Error: {errors[i-1]:.6f})')
plt.plot(x_range, func_exp(x_range, *popt_exp), label=f'Exponencial (Error: {error_exp:.6f})')
plt.plot(x_range, func_pow(x_range, *popt_pow), label=f'Potencial (Error: {error_pow:.6f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por mínimos cuadrados - Ejercicio 2')
plt.legend()
plt.grid(True)
plt.show()

for i, poly in enumerate(polys, 1):
    print(f"Polinomio de grado {i}: {poly}")
    print(f"Error: {errors[i-1]:.6f}")
print(f"Función exponencial: {popt_exp[1]:.6f} * e^({popt_exp[0]:.6f}x)")
print(f"Error: {error_exp:.6f}")
print(f"Función potencial: {popt_pow[1]:.6f} * x^{popt_pow[0]:.6f}")
print(f"Error: {error_pow:.6f}")
```

## Resolución del Ejercicio 3

La ecuación de la recta por mínimos cuadrados para los datos de puntuación ACT y promedio de puntos se puede encontrar aplicando las fórmulas:

$$b = \frac{n\sum_{i=1}^n x_iy_i - \sum_{i=1}^n x_i\sum_{i=1}^n y_i}{n\sum_{i=1}^n x_i^2 - \left(\sum_{i=1}^n x_i\right)^2}$$
$$a = \frac{\sum_{i=1}^n y_i - b\sum_{i=1}^n x_i}{n}$$

Donde:
- $x_i$ es la puntuación ACT
- $y_i$ es el promedio de puntos
- $n$ es el número de datos $(20)$

Reemplazando los valores, obtenemos:

$$b = \frac{20(1,443.21) - (548)(74.04)}{20(15,264) - (548)^2} = 0.0822$$
$$a = \frac{74.04 - 0.0822(548)}{20} = 2.6$$

Por lo tanto, la ecuación de la recta por mínimos cuadrados es:
$y = 2.6 + 0.0822x$

```python
# Ejercicio 3
print("\nEjercicio 3:")
act_scores = [28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 27, 29, 21, 28, 28, 26, 30, 24]
gpa_scores = [3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03, 3.75, 3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81]

coef = np.polyfit(act_scores, gpa_scores, 1)
poly = np.poly1d(coef)

plt.figure(figsize=(10, 6))
plt.scatter(act_scores, gpa_scores, color='blue', label='Datos')
plt.plot(act_scores, poly(act_scores), color='red', label='Recta de regresión')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.title('Relación entre puntuación ACT y promedio de puntos')
plt.legend()
plt.grid(True)
plt.show()

print(f"Ecuación de la recta: y = {coef[0]:.4f}x + {coef[1]:.4f}")
```

## Resolución del Ejercicio 4

$$a = \frac{\sum_{i=1}^n y_i - b\sum_{i=1}^n x_i}{n}$$

Donde:
- $x_i$ es el peso promedio
- $y_i$ es el porcentaje de presentación
- $n$ es el número de datos $(5)$

Reemplazando los valores, obtenemos:

$$b = \frac{5(1,109.6) - (19,700)(28.3)}{5(97,520,000) - (19,700)^2} = -0.000000426$$

Por lo tanto, la ecuación de la recta por mínimos cuadrados es:
$y = 5.66 - 0.000000426x$
